##Function Mapping and lambda
lambda function is called by its assigned variable as the function doesn't have a name of its own

```
frame = pd.DataFrame(np.random.randn(4,3), columns = list('bde'),
                    index = ['Utah', 'Columbo', 'Texas', 'Oregon'])
frame
```

```
frame.max()       #gives back the max value in each column
```

```
frame.min()
```

```
frame.abs()
```

```
f = lambda x: x.max() - x.min()
frame.apply(f)
```

```
f = lambda x: x ** 2
frame.apply(f)
```

##Sorting and Ranking

```
obj = pd.Series(range(4), index = [1,9,3,7])
obj
```

```
obj.sort_index(ascending = False)
```

```
frame = pd.DataFrame(np.arange(8).reshape((2,4)),
                     index = ['three','one'],
                     columns = ['d','e','a','c'])
frame
```

```
frame.sort_index()
```

```
frame = pd.DataFrame(np.arange(8, 0 , -1).reshape((2,4)),
                     index = ['three','one'],
                     columns = ['d','e','a','c'])
frame
```

```
frame.sort_values(by = 'a')
```

#Dictionary Comprehension

```
import pandas_datareader.data as web
```

```
all_data = { ticker: web.get_data_yahoo(ticker)
           for ticker in ['AAPL','IBM','MSFT','GOOG']}
all_data
```

```
price = pd.DataFrame({ticker:data['Adj Close']
                     for ticker, data in all_data.items()})
price
```

```
volume = pd.DataFrame({ticker : data['Volume']
                       for ticker, data in all_data.items()})
```

```
returns = price.pct_change()
returns
```

```
returns['MSFT'].corr(returns['IBM'])
```

```
returns['MSFT'].cov(returns['IBM'])
```

```
returns.corr()
```

```
returns.corrwith(returns.IBM)
```
